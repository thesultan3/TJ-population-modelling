```python
def filter_nyc_zip_codes(census_df):
    if "geoid" not in census_df.columns:
      raise ValueError("Column 'geoid' not found in the dataset. Check column names.")

    # remove everything before "US" in "geoid"
    census_df["geoid"] = census_df["geoid"].str.replace(r".*US", "", regex=True)
    census_df["geoid"] = census_df["geoid"].astype(str)
    print(f"Rows before filtering: {len(census_df)}")
    
    # Filter only valid NYC ZIP codes
    filtered_df = census_df[census_df["geoid"].isin(nyc_zip_codes)]
    print(f"Rows after filtering: {len(filtered_df)}")
    return filtered_df
```

```python
# clean col names
def clean_col_names(df):
  cols = df.columns = (
      df.columns
      .str.lower()
      .str.replace("!!", "_", regex=False) 
      .str.replace(" ", "_", regex=False)  
      .str.replace(r"[^\w_]", "", regex=True)
      .str.replace("estimate_total_", "", regex=False) 
  )
  return cols

def set_df_columns(df):
  df.columns = df.iloc[1]
  df = df.drop([0, 1]).reset_index(drop=True)
  df.rename(columns={"Geography": "geoid"}, inplace=True)
  df = filter_nyc_zip_codes(df)
  df = df.reset_index(drop=True)
  return df
```

```python
import pandas as pd
import geopandas as gpd
from shapely import wkt

# get datasets and setup geodataframes

ped_df = pd.read_csv('/content/Bi-Annual_Pedestrian_Counts_20250303.csv')
# take the average count from the last year's data
count_cols = ['June24_AM', 'June24_MD', 'June24_PM', 'Oct24_AM', 'Oct24_MD', 'Oct24_PM']
ped_df['avg_ped'] = ped_df[count_cols].mean(axis=1)
ped_df = ped_df[['the_geom', 'avg_ped']]
# Convert from WKT to geometry objects
ped_df['geometry'] = ped_df['the_geom'].apply(wkt.loads)
ped_df.drop(columns=['the_geom'], inplace=True)
# Create GeoDataFrame for ped count locations using geometric data
ped_gdf = gpd.GeoDataFrame(ped_df, geometry='geometry', crs="EPSG:4326")


# # this was inspired by two sigma's use of subway turnstile usage to estimate foot traffic - time permitting I could implement this in combination with other mapping of the turnstile identifiers to location (joining with other nyc station datasets)
# # """
# # Subway Turnstile Usage 2022
# # turnstile_df = pd.read_csv(turnstile_url)
# # Sum 2022 entries+exits per station (assuming columns like 'station', 'entries', 'exits')
# turnstile_df.columns = turnstile_df.columns.str.lower().str.replace(" ","", regex=False)
# print(turnstile_df.columns)
# turnstile_df['total_usage'] = turnstile_df['entries'] + turnstile_df['exits']
# print(turnstile_df.head)
# station_usage = turnstile_df.groupby('station_name')['total_usage'].sum().reset_index()

# MTA Subway Stations
stations_df = pd.read_csv('/content/MTA_Subway_Stations.csv')
stations_df = stations_df[['Stop Name', 'GTFS Longitude', 'GTFS Latitude']]
stations_gdf = gpd.GeoDataFrame(stations_df, geometry=gpd.points_from_xy(stations_df['GTFS Longitude'], stations_df['GTFS Latitude']), crs="EPSG:4326")

# Traffic Volume Counts
traffic_df = pd.read_csv('/content/Automated_Traffic_Volume_Counts.csv')
# Trader Joe's opening hours are usually 8am-9pm so we will only look at traffic between 7am-9pm
traffic_df = traffic_df[(traffic_df['HH'] >= 7) & (traffic_df['HH'] <= 20)]

# Aggregate traffic data directly in one step
traffic_df = traffic_df.groupby('RequestID').agg(
    total_vol_7am_8pm=('Vol', 'sum'),  # Total volume from 7 AM - 8 PM
    count=('Vol', 'count'),  # Number of records for that RequestID
    WktGeom=('WktGeom', 'first')  # Keep the WktGeom (it's unique by RequestID)
).reset_index()

# Calculate average traffic count per record to avoid oversampling bias
traffic_df['avg_count_7am_8pm'] = traffic_df['total_vol_7am_8pm'] / traffic_df['count']
traffic_df['geometry'] = traffic_df['WktGeom'].apply(wkt.loads)
traffic_df.drop(columns=['WktGeom'], inplace=True)
traffic_df = gpd.GeoDataFrame(traffic_df, geometry='geometry', crs="EPSG:4326")

# Parking Facilities
parking_df = pd.read_csv('/content/Licensed_Parking_Lots.csv')
parking_df = parking_df[['License Number','Latitude', 'Longitude']]
parking_df = gpd.GeoDataFrame(parking_df, geometry=gpd.points_from_xy(parking_df['Longitude'], parking_df['Latitude']), crs="EPSG:4326")

# Grocery Stores (Retail Food Stores)
stores_df = pd.read_csv('/content/Retail_Food_Stores.csv')

# select which retail stores we want to keep
chain_names = [
    "WHOLE FOODS MARKET GROUP INC",
    "WEGMANS FOOD MARKETS INC",
    "TRADER JOES EAST INC",
    "ALDI INC",
    "TARGET CORPORATION",
    "WAL-MART STORES EAST LP"
]

stores_df = stores_df[stores_df['Entity Name'].str.upper().isin(chain_names)].reset_index()

print(stores_df)
store_counts = stores_df['Entity Name'].value_counts()

stores_df['geometry'] = stores_df['Georeference'].apply(wkt.loads)

stores_gdf = gpd.GeoDataFrame(stores_df, geometry='geometry', crs="EPSG:4326")
stores_gdf = stores_gdf[['Entity Name', 'Square Footage', 'geometry']]
```

```python
# NYC has many zip codes, often assigned to small areas/buildings. This dataset simplifies this into larger (real?) zip codes that act as better measures, it was used for modelling zip based covid values during the pandemic.
zip_df = pd.read_csv('/content/Modified_Zip_Code_Tabulation_Areas__MODZCTA__20250303.csv')

# Convert the 'the_geom' column (WKT) into geometry objects
zip_df['geometry'] = zip_df['the_geom'].apply(wkt.loads)

# Create a GeoDataFrame from zip_df
zip_gdf = gpd.GeoDataFrame(zip_df, geometry='geometry', crs="EPSG:4326")
zip_gdf.rename(columns={'MODZCTA': 'ZIP'}, inplace=True)
zip_gdf['ZIP'] = zip_gdf['ZIP'].astype(str)

# this is unfortunately very sparse (only covers 59 zip codes) so it is likely an unreliable indicator
# Pedestrian Counts (ped_gdf is already created)
ped_zip = gpd.sjoin(ped_gdf, zip_gdf[['ZIP', 'geometry']], how='left', predicate='intersects')
# Aggregate average pedestrian count per ZIP
ped_by_zip = ped_zip.groupby('ZIP').agg({'avg_ped': 'mean'}).reset_index()

# Traffic Data
traffic_zip = gpd.sjoin(traffic_df, zip_gdf[['ZIP', 'geometry']], how='left', predicate='intersects')
# Aggregate sum total volume and average the per-record average for each ZIP
traffic_by_zip = traffic_zip.groupby('ZIP').agg({
    'total_vol_7am_8pm': 'sum',
    'avg_count_7am_8pm': 'mean'
}).reset_index()

# Parking data
parking_zip = gpd.sjoin(parking_df, zip_gdf[['ZIP', 'geometry']], how='left', predicate='intersects')
# count the number of parking facilities per ZIP
parking_by_zip = parking_zip.groupby('ZIP').size().reset_index(name='parking_count')

# Grocery store locations data
stores_zip = gpd.sjoin(stores_gdf, zip_gdf[['ZIP', 'geometry']], how='left', predicate='intersects')

stores_by_zip = stores_zip.groupby(['ZIP', 'Entity Name']).size().unstack(fill_value=0).reset_index()
stores_by_zip = stores_by_zip.rename(columns={
    "WHOLE FOODS MARKET GROUP INC": "whole_foods_count",
    "WEGMANS FOOD MARKETS INC": "wegmans_count",
    "TRADER JOES EAST INC": "trader_joes_count",
    "ALDI INC": "aldi_count",
    "TARGET CORPORATION": "target_count",
    "WAL-MART STORES EAST LP": "walmart_count"
})

# (E) Grocery Store Locations (stores_gdf is already a GeoDataFrame)
stations_zip = gpd.sjoin(stations_gdf, zip_gdf[['ZIP', 'geometry']], how='left', predicate='intersects')
# Aggregate: count stores per ZIP and average square footage
stations_by_zip = stations_zip.groupby('ZIP').size().reset_index(name='station_count')
```

```python
gdf = gpd.read_file("final_combined_data.geojson")

# ===============================
# 1. Population by Age Groups
# ===============================
# Total population under 20 = Sum ages: under 5, 5-9, 10-14, 15-19
gdf['pop_under_20'] = (
    gdf['total_population_age_under_5_years'] +
    gdf['total_population_age_5_to_9_years'] +
    gdf['total_population_age_10_to_14_years'] +
    gdf['total_population_age_15_to_19_years']
)

# Total population 20-30 = Sum ages: 20-24, 25-29
gdf['pop_20_30'] = (
    gdf['total_population_age_20_to_24_years'] +
    gdf['total_population_age_25_to_29_years']
)

# Total population 30-45 = Sum ages: 30-34, 35-39, 40-44
gdf['pop_30_45'] = (
    gdf['total_population_age_30_to_34_years'] +
    gdf['total_population_age_35_to_39_years'] +
    gdf['total_population_age_40_to_44_years']
)

# Total population 45-60 = Sum ages: 45-49, 50-54, 55-59
gdf['pop_45_60'] = (
    gdf['total_population_age_45_to_49_years'] +
    gdf['total_population_age_50_to_54_years'] +
    gdf['total_population_age_55_to_59_years']
)

# Total population 60+ = Sum ages: 60-64, 65-69, 70-74, 75-79, 80-84, 85+
gdf['pop_60_plus'] = (
    gdf['total_population_age_60_to_64_years'] +
    gdf['total_population_age_65_to_69_years'] +
    gdf['total_population_age_70_to_74_years'] +
    gdf['total_population_age_75_to_79_years'] +
    gdf['total_population_age_80_to_84_years'] +
    gdf['total_population_age_85_years_and_over']
)

# ===============================
# 2. Income Groups
# ===============================
# Low income: income under 40000, sum the following columns
gdf['low_income'] = (
    gdf['less_than_10000'] +
    gdf['10000_to_14999'] +
    gdf['15000_to_19999'] +
    gdf['20000_to_24999'] +
    gdf['25000_to_29999'] +
    gdf['30000_to_34999'] +
    gdf['35000_to_39999']
)

# Middle income: 40000-125000, sum these columns
gdf['middle_income'] = (
    gdf['40000_to_44999'] +
    gdf['45000_to_49999'] +
    gdf['50000_to_59999'] +
    gdf['60000_to_74999'] +
    gdf['75000_to_99999'] +
    gdf['100000_to_124999']
)

# High income: 125000+, sum these columns
gdf['high_income'] = (
    gdf['125000_to_149999'] +
    gdf['150000_to_199999'] +
    gdf['200000_or_more']
)

# ===============================
# 3. Education: Bachelors Degree (or Higher)
# ===============================
# Sum: bachelors_degree, masters_degree, professional_school_degree, doctorate_degree
gdf['bachelors_or_higher'] = (
    gdf['bachelors_degree'] +
    gdf['masters_degree'] +
    gdf['professional_school_degree'] +
    gdf['doctorate_degree']
)

gdf.to_csv("final_combined_data_with_new_columns.csv", index=False)
gdf.to_file('final_combined_data_with_new_columns.geojson', driver='GeoJSON')
```

```python
%%writefile app.py
import streamlit as st
import geopandas as gpd
import folium
from streamlit_folium import st_folium
import pandas as pd
from shapely import wkt

# Load into GeoDataFrame
gdf = gpd.read_file("final_combined_data_with_new_columns.geojson")

st.title("Interactive Store Suitability Heatmap")
st.write("Adjust the weights below to compute a composite suitability score for each ZIP code.")

for col in [
    'total_population', 'median_household_income',
    'pop_under_20','pop_20_30','pop_30_45','pop_45_60','pop_60_plus',
    'low_income','middle_income','high_income',
    'bachelors_or_higher',
    'parking_count',
    'commute_total',
    'hispanic_or_latino','not_hispanic_or_latino','white','black_or_african_american', 'asian'
          ]:
    gdf[col] = pd.to_numeric(gdf[col], errors='coerce').fillna(0) # fill missing values

# list of features
factors = [
    'total_population', 'median_household_income',
    'pop_under_20','pop_20_30','pop_30_45','pop_45_60','pop_60_plus',
    'low_income','middle_income','high_income',
    'bachelors_or_higher',
    'parking_count',
    'commute_total',
    'hispanic_or_latino','not_hispanic_or_latino','white','black_or_african_american', 'asian',
          ]

st.subheader("Set Weights for Each Factor")
# make sliders 0 to 1 for each feature, start at 0
weights = {}
for factor in factors:
    weights[factor] = st.slider(f"Weight for {factor}", 0.0, 1.0, 0.0, step=0.01)

def normalize(series):
    min_val = series.min()
    max_val = series.max()
    if max_val - min_val == 0:
        return series * 0
    return (series - min_val) / (max_val - min_val)

# normalise
for factor in factors:
    norm_col = f"norm_{factor}"
    gdf[norm_col] = normalize(gdf[factor])

# calc composite score using norm vals
total_weight = sum(weights.values())
if total_weight > 0:
    gdf['score'] = sum(weights[factor] * gdf[f"norm_{factor}"] for factor in factors) / total_weight
else:
    gdf['score'] = 0

# Folium map centered on NYC
m = folium.Map(location=[40.7128, -74.0060], zoom_start=10, tiles='cartodbpositron')

# Add a choropleth layer for the composite score
choropleth = folium.Choropleth(
    geo_data=gdf.to_json(),
    data=gdf,
    columns=['ZIP', 'score'],
    key_on='feature.properties.ZIP',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Composite Suitability Score (0-1)'
).add_to(m)

folium.GeoJsonTooltip(
    fields=factors,
    localize=True
).add_to(choropleth.geojson)

# make pointers on the map for the stores
stores_df = pd.read_csv("Retail_Food_Stores.csv")
stores_df['Entity Name'] = stores_df['Entity Name'].str.upper()

tj_df = stores_df[stores_df['Entity Name'].str.contains("TRADER JOE", na=False)]
wf_df = stores_df[stores_df['Entity Name'].str.contains("WHOLE FOODS", na=False)]
wg_df = stores_df[stores_df['Entity Name'].str.contains("WEGMAN", na=False)]

tj_df['geometry'] = tj_df['Georeference'].apply(wkt.loads)
wf_df['geometry'] = wf_df['Georeference'].apply(wkt.loads)
wg_df['geometry'] = wg_df['Georeference'].apply(wkt.loads)

tj_gdf = gpd.GeoDataFrame(tj_df, geometry='geometry', crs="EPSG:4326")
wf_gdf = gpd.GeoDataFrame(wf_df, geometry='geometry', crs="EPSG:4326")
wg_gdf = gpd.GeoDataFrame(wg_df, geometry='geometry', crs="EPSG:4326")

# add score to TJ stores for other computations (mean, std etc)
tj_join = gpd.sjoin(tj_gdf, gdf[['ZIP', 'score', 'geometry']], how='left', predicate='within')

for idx, row in tj_join.iterrows():
    folium.CircleMarker(
        location=[row.geometry.y, row.geometry.x],
        radius=1,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.8,
        popup=f"Trader Joe's: {row['Entity Name']}<br>Score: {row['score']:.2f}"
    ).add_to(m)

for idx, row in wf_gdf.iterrows():
    folium.CircleMarker(
        location=[row.geometry.y, row.geometry.x],
        radius=3,
        color='green',
        fill=True,
        fill_color='green',
        fill_opacity=0.8,
        popup=f"Whole Foods: {row['Entity Name']}"
    ).add_to(m)

for idx, row in wg_gdf.iterrows():
    folium.CircleMarker(
        location=[row.geometry.y, row.geometry.x],
        radius=1,
        color='purple',
        fill=True,
        fill_color='purple',
        fill_opacity=0.8,
        popup=f"Wegmans : {row['Entity Name']}"
    ).add_to(m)

# setup station data, can comment this out (there are a lot of stations can get in the way of the big picture)
stations_df = pd.read_csv("MTA_Subway_Stations.csv")
stations_df = stations_df[['Stop Name', 'GTFS Latitude', 'GTFS Longitude']]
stations_gdf = gpd.GeoDataFrame(
    stations_df,
    geometry=gpd.points_from_xy(stations_df['GTFS Longitude'], stations_df['GTFS Latitude']),
    crs="EPSG:4326"
)

for idx, row in stations_gdf.iterrows():
    folium.CircleMarker(
        location=[row.geometry.y, row.geometry.x],
        radius=0.2,
        color='grey',
        fill=True,
        fill_color='red',
        fill_opacity=0.8,
        popup=f"Station: {row['Stop Name']}"
    ).add_to(m)

# calc TJ stats according to the current scores (from current equation)
avg_score = tj_join['score'].mean()
min_score = tj_join['score'].min()
max_score = tj_join['score'].max()
std_score = tj_join['score'].std()

st.write("Trader Joe's Composite Score Statistics:")
st.write(f"Average: {avg_score:.2f}")
st.write(f"Minimum: {min_score:.2f}")
st.write(f"Maximum: {max_score:.2f}")
st.write(f"Standard Deviation: {std_score:.2f}")


# -------------------------------
# Highlight ZIP 11215 with a custom green style
# -------------------------------
target_zip = "11215"

# for making final visualisation highlighting and marker (for main plot)
# highlight_layer = folium.GeoJson(
#     gdf[gdf['ZIP'] == target_zip].to_json(),
#     style_function=lambda feature: {
#         'fillColor': 'green',
#         'color': 'green',
#         'weight': 2,
#         'fillOpacity': 0.5
#     }
# )
# highlight_layer.add_to(m)

# recommended_location = [40.6665, -73.9950]  # Example coordinates
# folium.Marker(
#     location=recommended_location,
#     popup="Recommended Location in ZIP 11215",
#     icon=folium.Icon(color='green', icon='ok-sign')
# ).add_to(m)

# Display the map in Streamlit
st_data = st_folium(m, width=700, height=500)
```
